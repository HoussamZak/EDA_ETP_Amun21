21 Shares Crypto Basket Index (Tracks BTC ETH SOL DOT ADA as investments results of an index composed of top 5 crypto currencies ranked by 2050 market capitalization.)
To compare all ETPs, refer to our list here: https://docs.google.com/spreadsheets/d/1ora88Kh1GjfPjd3lK96VjUR8WbNDvdXVkw1KFDcRrlY/edit?usp=sharing

Dates FEB 21 2017 => FEB 20 2022    

```python
import pandas as pd
import yfinance as yf

#List of altcoin pairs in the ETP
#For Polkadot, I wrote DOT-CAD because the USD pair gives no data.

crypto_basket_five = ['BTC-USD', 'ETH-USD', 'SOL-USD', 'DOT-CAD', 'ADA-USD']
crypto_basket_five_temp = yf.download(crypto_basket_five, start = "2017-02-21", end = "2022-02-20", data_source = "yahoo")
crypto_basket_five_temp.drop(columns = ['Volume', 'Open', 'Close','Low', 'High'], inplace= True)

crypto_basket_five_temp.tail()
```

```python
#Loading the btc_metrics file

btc_metrics = pd.read_csv(r'C:\Users\Fluffy\Desktop\Projects Coding\EDA_ETP_Amun21\btc_metrics.csv')
# btc_metrics = btc_metrics.drop(columns=['Unnamed: 0'],inplace= True)

# #Creating a Date column to merge with the btc_metrics DF
btc_metrics['Date'] = crypto_basket_five_temp.index
btc_metrics.drop(columns=['Unnamed: 0'], inplace=True)
btc_metrics = btc_metrics[['Date','price','volume','bitfinex_TradVol','coinbase_TradVol','gemini_TradVol','kraken_TradVol','bitfinex_DELTA_TradeVolume','gemini_DELTA_TradeVolume'	, 'kraken_DELTA_TradeVolume', 'coinbase_DELTA_TradeVolume',	'bitfinex_Vola'	,'coinbase_VOLA', 'gemini_VOLA','kraken_VOLA','bitfinex_B/A_Sp','coinbase_B/A_Sp','gemini_B/A_Sp','kraken_B/A_Sp']]
btc_metrics
```

Merging Bitcoin Metrics & Price Data of Altcoins included 21Shares Crypto Basket ETP(BTC,ETH,ADA,SOL,DOT)

```python
#Merging bitcoin metrics & Altcoin prices
btc_metrics_and_crypto_basket= pd.merge(crypto_basket_five_temp, btc_metrics, how='right', on = ['Date', 'Date'])

#We made sure to drop the DATE column as it was giving an erroneous correlation matrix!!! 
btc_metrics_and_crypto_basket.drop(columns=['Date'], inplace = True)

#Renaming columns of our DF for better readability
btc_metrics_and_crypto_basket.columns =['ADA-USD','BTC-USD'	,'DOT-CAD'	,'ETH-USD','SOL-USD', 'price','volume','bitfinex_TradVol',
'coinbase_TradVol','gemini_TradVol','kraken_TradVol', 'bitfinex_DELTA_TradeVolume','gemini_DELTA_TradeVolume',   'kraken_DELTA_TradeVolume','coinbase_DELTA_TradeVolume','bitfinex_Vola','coinbase_VOLA','gemini_VOLA',
'kraken_VOLA','bitfinex_B/A_Sp','coinbase_B/A_Sp','gemini_B/A_Sp','kraken_B/A_Sp']

#Preview
btc_metrics_and_crypto_basket.tail()
```

Building our correlation model then Correlation Heatmap !

```python
btc_metrics_and_crypto_basket_columns = btc_metrics_and_crypto_basket.columns
btc_metrics_and_crypto_basket_corr = pd.DataFrame(btc_metrics_and_crypto_basket, columns=btc_metrics_and_crypto_basket_columns)
btc_metrics_and_crypto_basket_corr_df = btc_metrics_and_crypto_basket_corr.corr()
btc_metrics_and_crypto_basket_corr_df
```

```python
import numpy as np
import seaborn as sns

from matplotlib import pyplot as plt

mask = np.triu(np.ones_like(btc_metrics_and_crypto_basket_corr_df, dtype=np.bool))
f, ax = plt.subplots(figsize=(15,15))

cmap ="RdBu"
sns.heatmap(
    btc_metrics_and_crypto_basket_corr_df,
    mask = mask,
    cmap = cmap,
    center = 0,
    square = True,
    linewidths = 0.5,
    cbar_kws ={"shrink":0.5}
)
```

Obviously, it is difficult to spot quickly obvious/high correlations simply by looking above.
So let's filter weak correlations from our matrix, below, focusing on correlation >= 0.8 and <= -0.8 !

```python
dfCorr =  btc_metrics_and_crypto_basket_corr_df.corr()
filteredDf = dfCorr[((dfCorr >= .8) | (dfCorr <= -.8)) & (dfCorr !=1.000)]
plt.figure(figsize=(30,10))
sns.heatmap(filteredDf, annot=True, cmap="RdBu")
plt.show()
```

```python
def corrFilter(x: pd.DataFrame, bound: float):
    xCorr = x.corr()
    xFiltered = xCorr[((xCorr >= bound) | (xCorr <= -bound)) & (xCorr !=1.000)]
    xFlattened = xFiltered.unstack().sort_values().drop_duplicates()
    return xFlattened

pd.set_option('display.max_rows', None)

(corrFilter( btc_metrics_and_crypto_basket_corr_df, .8))
```
