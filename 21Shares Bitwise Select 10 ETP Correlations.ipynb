```python
import pandas as pd
import yfinance as yf

#List of altcoin pairs in the ETP
#For Polkadot, I wrote DOT-CAD because the USD pair gives no data.
bitwise_select_top10 = ['BTC-USD', 'ETH-USD','AVAX-USD', 'SOL-USD', 'ADA-USD', 'BCH-USD', 'LTC-USD', 'MATIC-USD','ATOM-USD', 'DOT-CAD']
bitwise_select_top10_temp = yf.download(bitwise_select_top10, start = "2017-02-21", end = "2022-02-20", data_source = "yahoo")
bitwise_select_top10_temp.drop(columns = ['Volume', 'Open', 'Close','Low', 'High'], inplace= True)

#Merging with btc_data
# btc_metrics_and_15 = pd.merge(btc_metrics, fifteen_ficas_active_crypto_temp, how = 'right', on =['Date', 'Date'])

bitwise_select_top10_temp
```

```python
#Loading the btc_metrics file

btc_metrics = pd.read_csv(r'C:\Users\Fluffy\Desktop\Projects Coding\EDA_ETP_Amun21\btc_metrics.csv')
# btc_metrics = btc_metrics.drop(columns=['Unnamed: 0'],inplace= True)

# #Creating a Date column to merge with the btc_metrics DF
btc_metrics['Date'] = bitwise_select_top10_temp.index
btc_metrics.drop(columns=['Unnamed: 0'], inplace=True)
btc_metrics = btc_metrics[['Date', 'price','volume','bitfinex_TradVol','coinbase_TradVol','gemini_TradVol','kraken_TradVol','bitfinex_DELTA_TradeVolume','gemini_DELTA_TradeVolume'	, 'kraken_DELTA_TradeVolume', 'coinbase_DELTA_TradeVolume',	'bitfinex_Vola'	,'coinbase_VOLA', 'gemini_VOLA','kraken_VOLA','bitfinex_B/A_Sp','coinbase_B/A_Sp','gemini_B/A_Sp','kraken_B/A_Sp']]
btc_metrics
```

Merging Bitcoin Metrics & Altcoins included in 21SHARES BITWISE SELECT 10 ETP

```python
#Merging bitcoin metrics & Altcoin prices
btc_metrics_and_bitwisetop10 = pd.merge(bitwise_select_top10_temp, btc_metrics, how='right', on = ['Date', 'Date'])
btc_metrics_and_bitwisetop10.drop(columns=['Date'], inplace = True)


#Renaming columns of our DF for better readability

btc_metrics_and_bitwisetop10.columns =[                         ('ADA-USD'),
          ('ATOM-USD'),    ('AVAX-USD'),
           ('BCH-USD'),     ('BTC-USD'),
           ('DOT-CAD'),     ('ETH-USD'),
           ('LTC-USD'),   ('MATIC-USD'),
           ('SOL-USD'),                      'price',
                           'volume',           'bitfinex_TradVol',
                 'coinbase_TradVol',             'gemini_TradVol',
                   'kraken_TradVol', 'bitfinex_DELTA_TradeVolume',
         'gemini_DELTA_TradeVolume',   'kraken_DELTA_TradeVolume',
       'coinbase_DELTA_TradeVolume',              'bitfinex_Vola',
                    'coinbase_VOLA',                'gemini_VOLA',
                      'kraken_VOLA',            'bitfinex_B/A_Sp',
                  'coinbase_B/A_Sp',              'gemini_B/A_Sp',
                    'kraken_B/A_Sp']


btc_metrics_and_bitwisetop10
```

Building our correlation model then Correlation Heatmap !

```python

bitwisetop10_columns = btc_metrics_and_bitwisetop10.columns
btc_metrics_and_bitwisetop10_corr = pd.DataFrame(btc_metrics_and_bitwisetop10, columns=bitwisetop10_columns)
btc_metrics_and_bitwisetop10_corr_df = btc_metrics_and_bitwisetop10_corr.corr()
btc_metrics_and_bitwisetop10_corr_df
```

```python
import numpy as np
import seaborn as sns

from matplotlib import pyplot as plt

mask = np.triu(np.ones_like(btc_metrics_and_bitwisetop10_corr_df, dtype=np.bool))
f, ax = plt.subplots(figsize=(15,15))

cmap ="RdBu"
sns.heatmap(
    btc_metrics_and_bitwisetop10_corr_df,
    mask = mask,
    cmap = cmap,
    center = 0,
    square = True,
    linewidths = 0.5,
    cbar_kws ={"shrink":0.5}
)
```

Obviously, it is difficult to spot quickly obvious/high correlations simply by looking above.
So let's filter weak correlations from our matrix, below, focusing on correlation >= 0.8 and <= -0.8 !

```python
dfCorr = btc_metrics_and_bitwisetop10_corr_df.corr()
filteredDf = dfCorr[((dfCorr >= .8) | (dfCorr <= -.8)) & (dfCorr !=1.000)]
plt.figure(figsize=(30,10))
sns.heatmap(filteredDf, annot=True, cmap="RdBu")
plt.show()
```

The observations & conclusions for 21Shares Bitwise Select 10 ETP can be assumed for 21 Shares Crypto Basket 10 ETP which is a similar ETP that rebalances quarterly.
