Building our correlation model then our Correlation Heatmap!

```python
btc_metrics_and_sygnum_winners_columns = btc_metrics_and_sygnum_winners.columns
btc_metrics_and_sygnum_winners_corr = pd.DataFrame(btc_metrics_and_sygnum_winners, columns=btc_metrics_and_sygnum_winners_columns)
btc_metrics_and_sygnum_winners_corr_df = btc_metrics_and_sygnum_winners_corr.corr()
btc_metrics_and_sygnum_winners_corr_df
```

```python
import numpy as np
import seaborn as sns

from matplotlib import pyplot as plt

mask = np.triu(np.ones_like(btc_metrics_and_sygnum_winners_corr_df, dtype=np.bool))
f, ax = plt.subplots(figsize=(15,15))

cmap ="RdBu"
sns.heatmap(
    btc_metrics_and_sygnum_winners_corr_df,
    mask = mask,
    cmap = cmap,
    center = 0,
    square = True,
    linewidths = 0.5,
    cbar_kws ={"shrink":0.5}
)
```

Obviously, it is difficult to spot quickly obvious/high correlations simply by looking above.
So let's filter weak correlations from our matrix, below, focusing on correlation >= 0.8 and <= -0.8 !

```python
dfCorr =  btc_metrics_and_sygnum_winners_corr_df.corr()
filteredDf = dfCorr[((dfCorr >= .8) | (dfCorr <= -.8)) & (dfCorr !=1.000)]
plt.figure(figsize=(30,10))
sns.heatmap(filteredDf, annot=True, cmap="RdBu")
plt.show()
```

List of strongest Correlations

```python
def corrFilter(x: pd.DataFrame, bound: float):
    xCorr = x.corr()
    xFiltered = xCorr[((xCorr >= bound) | (xCorr <= -bound)) & (xCorr !=1.000)]
    xFlattened = xFiltered.unstack().sort_values().drop_duplicates()
    return xFlattened

pd.set_option('display.max_rows', None)

(corrFilter( btc_metrics_and_sygnum_winners_corr_df, .8))
```
